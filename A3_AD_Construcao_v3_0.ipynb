### Adicionando uma coluna identificadora em cada conjunto

Para conseguirmos identificar de qual conjunto de dados uma determinada instância veio, iremos adicionar um atributo novo em cada dataset para informar a origem das instâncias.

Resolveremos esse problema utilizando a função "np.full" da biblioteca "numpy".

Ela é capaz de gerar um vetor preenchido com um determinado valor.

Criaremos 4 vetores e os preencheremos com os valores: "Amazon Prime", "Netflix", "Disney+" e "Hulu".

Esses vetores se chamarão: "identificador_amazon", "identificador_disney_plus", "identificador_netflix" e "identificador_hulu".

```python
identificador_disney_plus = np.full(len(dados_disney_plus), "Disney+")
```

Iremos nomear a coluna que guardará os vetores identificadores de "dataset_de_origem".

```python
dados_amazon["dataset_de_origem"] = identificador_amazon
```

```python
dados_disney_plus["dataset_de_origem"] = identificador_disney_plus
```

```python
dados_netflix["dataset_de_origem"] = identificador_netflix
```

```python
dados_hulu["dataset_de_origem"] = identificador_hulu
```

Por último, vamos verificar se todos os datasets realmente acolheram uma coluna que identifica suas amostras.

Para unir os conjuntos de dados, iremos utilizar o método "concat" da biblioteca "pandas". O argumento "axis=0" indica que queremos a união feita concatenando um conjunto embaixo do outro.

### Transformando a coluna "duration" em duas colunas: "duration" e "seasons"

Em nosso conjunto de dados, existe uma categoria chamada "type (tipo)" a qual pode possuir um de dois valores: Movie (Filme) ou TV Show (Série de TV), como demonstrado abaixo.

```python
dados_mesclados["type"].value_counts()
```

O problema disso é que TV Show e Movies não possuem seus valores na coluna "duration" medidos da mesma forma.

Movies tem duração apresentada em minutos.

TV Show tem sua duração apresentada em "Seasons (Temporadas)".

Diante desse cenário, iremos criar uma coluna apenas para duração em minutos e outra para quantidade de temporadas.

A coluna que guarda a quantidade de temporadas irá se chamar "seasons".

Para colocar nessa coluna apenas as temporadas, iremos aplicar uma expressão lambda na coluna "duration" para transformar em 0 todos os valores que não tiverem a substring "Season" neles. O resto dos valores, os quais são os valores que queremos, permanecerão os mesmos.

```python
dados_mesclados["seasons"] = dados_mesclados["duration"].apply(lambda x: 0 if not "Season" in x else x)
```

Verificando se a coluna foi adicionada:

```python
dados_mesclados.head()
```

```python
dados_mesclados["seasons"].value_counts()
```

A operação foi bem sucedida.

Agora, removeremos todos os valores que indicam quantidade de temporadas da coluna "duration" pois a mesma só irá registrar duração em minutos de filmes a partir de agora.

```python
dados_mesclados["duration"] = dados_mesclados["duration"].apply(lambda x: 0 if "Season" in x else x)
```

Verificando se a remoção foi bem sucedida:

```python
dados_mesclados["duration"].value_counts()
```

Podemos ver acima que o valor 0 aparecee 3490 vezes.

3490 é exatamente o número de amostras de "TV Show" que nosso conjunto de dados contém.

Isso significa que 3490 valores da coluna "duration", os quais eram números referentes a quantidades de temporadas, foram removidos dessa coluna.

### Removendo a substring "min" dos valores da coluna "duration"

Prosseguindo, iremos remover a substring "min", presente nos valores da coluna "duration"

Faremos isso para facilitar futuras operações envolvendo esses valores.

Utilizando o método "apply", criaremos uma expressão lambda a qual usará o método "replace" para substituir a subtring "min", presente nos valores, por uma string vazia.

Caso a expressão lambda encontre um valor 0 (ou seja, um valor que não contém a substring "min") ela o deixará como está.

```python
dados_mesclados["duration"] = dados_mesclados["duration"].apply(lambda x: 0 if x == 0 else x.replace(" min", ''))
```

Verificando se tudo ocorreu como esperado:

```python
dados_mesclados["duration"].value_counts()
```

Como é visível acima, a substring "min" sumiu dos valores.

Por fim, converteremos o tipo dessa coluna de para "inteiro".

```python
dados_mesclados["duration"] = dados_mesclados["duration"].astype("int")
```

### Removendo as substrings "Season" e "Seasons" dos valores da coluna "seasons"

Pelo mesmo motivo que removemos as substrings "min" na etapa anterior, iremos também remover "Season" e "Seasons", substrings presentes na coluna "seasons".

Utilizando o método "apply", criaremos uma expressão lambda a qual usará o método "replace" para substituir as subtrings "Season" e "Seasons", presente nos valores, por uma string vazia.

Caso a expressão lambda encontre um valor 0 (ou seja, um valor que não contém as substrings citadas) ela o deixará como está.

Primeiro, iremos remover a substring "Seasons":

```python
dados_mesclados["seasons"] = dados_mesclados["seasons"].apply(lambda x: 0 if x == 0 else x.replace("Seasons", ""))
```

Agora, removeremos a substring "Season":

```python
dados_mesclados["seasons"] = dados_mesclados["seasons"].apply(lambda x: 0 if x == 0 else x.replace("Season", ""))
```

Conferindo se foram removidas:

```python
dados_mesclados["seasons"].value_counts()
```

Também, converteremos o tipo dessa coluna para "inteiro".

```python
dados_mesclados["seasons"] = dados_mesclados["seasons"].astype("int")
```

### Transformando a coluna "type" em duas colunas: "movie" e "tv_series"

Ainda visando facilitar a visualização e operações futuras, iremos dividir o atributo "type" em duas colunas as quais guardarão apenas valores 0 ou 1.

A primeira coluna se chamará "movie" e guardará valores 1 caso a amostra seja um "Movie" e 0 caso seja uma "TV Show".

A segunda coluna se chamará "tv_series" e guardará valores 1 caso a amostra seja uma "TV Show" e 0 caso seja "Movie".

Criaremos elas através, também, de expressões lambda.

```python
dados_mesclados["movie"] = dados_mesclados["type"].apply(lambda x: 1 if "Movie" in x else 0)
```

```python
dados_mesclados["tv_series"] = dados_mesclados["type"].apply(lambda x: 1 if "TV" in x else 0)
```

Verificando se as operações funcionaram:

```python
dados_mesclados["movie"].value_counts()
```

```python
dados_mesclados["tv_series"].value_counts()
```

Considerando que tínhamos 3490 "TV Shows" no nosso conjunto de dados, percebemos acima que a coluna "tv_series" possui sua quantidade de valores iguais a 1 também em 3490.

Após isso, removeremos completamente a coluna "type" do DataFrame pois não a usaremos mais.

```python
dados_mesclados = dados_mesclados.drop("type", axis=1)
```

Conferindo os resultados:

```python
dados_mesclados.head()
```

A coluna "type" foi removida com sucesso.

### Transformando a coluna "date added" para a mesma salvar apenas o ano

Nesta última seção de modificação, alteraremos o formato dos dados guardados pela coluna "date_added".

Essa coluna registra o dia, mês e ano em que um filme ou série foi adicionado à plataforma de streaming a qual ele pertence.

Contudo, faremos ela registrar apenas o ano por dois fatores:

1. Não iremos utilizar o dia e mês de lançamento de obras em nossas análises.
2. Fazer essa coluna guardar apenas o ano facilitará os cálculos que faremos em cima dela.

Começaremos criando uma lista para guardar todos os anos de todas as datas em que as obras foram adicionadas à plataformas de streaming.

```python
lista_de_anos = []
```

Após isso, iteraremos por todas as linhas da coluna "date_added", coletaremos apenas os anos através da função "split", salvaremos o valor na variável "ano" e adicionaremos esse valor na lista de anos.

```python
for linha in dados_mesclados["date_added"]:
  mes, ano = linha.split(", ")
  lista_de_anos.append(int(ano))
```

Para transformar a coluna, trocaremos todos os valores dela pelos valores salvos na lista de anos.

```python
dados_mesclados["date_added"] = lista_de_anos
```

No final, alteraremos o nome da coluna para "year_added" a fim de facilitar o entendimento.

```python
dados_mesclados.rename(columns={"date_added": "year_added"}, inplace=True)
```
